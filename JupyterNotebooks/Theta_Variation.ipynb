# Investigation of a Simple Pendulum

The differential equation associated with a simple pendulum is:

\begin{equation}
\frac{d^2 \theta}{dt^2} = \frac{m g L_{cm}}{I_{pivot}} sin(\theta) \nonumber
\end{equation}

The so-called "small angle approximation" is often used, where we assume that $sin(\theta) \approx \theta$.  The first question is - how good an approximation is this?

```python
import numpy as np
import matplotlib.pyplot as plt

from P201_Functions import *

set_dark_mode(True)


theta  = np.array([(1+i)/100.0 for i in range(500)])
thetar = theta*np.pi/180.0
sintheta = np.sin(thetar)

diff = (thetar-sintheta)/sintheta

plt.plot(theta,diff)
plt.xlabel('Theta (degrees)')
plt.ylabel('(Theta - sin(Theta))/sin(Theta)')
```

The conclusion here would be, then, that if we wished to make measurements that were accurate to, say, 0.00005, we would need to keep $\theta < 1^\circ$.

However, a fuller treatment of the simple pendulum problem, including elliptic integrals, allows us to correct for the non-small-angle deviations!  

```python
correction = 1.0 + 1.0/16*thetar**2 + 11.0/3072.0*thetar**4 + 173.0/737820.0*thetar**6

plt.plot(theta,(correction-1)/0.375)
plt.xlabel('Theta (degrees)')
plt.ylabel('Correction Factor')
```

```python
plt.plot(theta,(((correction-1)/0.375-diff)/0.00005))
plt.xlabel('Theta (degrees)')
plt.ylabel('(Correction Factor-Diff)/0.00005')
```

Conclusion:  We can correct for the small angle approximation, and the resulting systematic uncertainty, after correction, should be about 0.012 the size of our desired error bar (0.00005) at a starting angle of 5 degrees!
